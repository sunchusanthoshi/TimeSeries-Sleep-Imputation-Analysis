```python
# Plotting the class distribution
class_counts = df['sleeping'].value_counts()
class_labels = class_counts.index

plt.figure(figsize=(6, 4))
plt.bar(class_labels, class_counts, color=['orange', 'blue'])
plt.xlabel('Sleeping')
plt.ylabel('Count')
plt.title('Class Distribution')
plt.xticks(class_labels, ['Not Sleeping', 'Sleeping'])
plt.show()
```

```python
df=pd.read_parquet('dataset.parquet')
df.head()
df=df[['timestamp','anglez','enmo','year','month','day','hour','minute','sleeping']]
# # df=pd.read_parquet(r'D:/CIS 550/child-mind-institute-detect-sleep-states/clean_train_par.parquet')
df.head()
# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
